```python
from sklearn import datasets
import pandas as pd
import seaborn
```

```python
iris = datasets.load_iris()
print(iris.target_names)
df = pd.DataFrame(iris.data,columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
```

```python
print(df)
```

```python
df['Species'] = iris.target
```

```python
df.Species.replace([0,1,2],['setosa','versicolor','virginica'],inplace=True)
display(df)
```

```python
df['Species'].value_counts().plot(kind='bar',title='Iris Species Distribution',color=['g','b','r'])
```

```python
df.plot(x='petal width (cm)',y='sepal width (cm)',kind='scatter',title='Petal width Vs Sepal width')
```

```python
df['petal length (cm)'].plot.density(color=['g','r','b','y'])
```
