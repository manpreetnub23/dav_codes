```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Name": [chr(x) for x in range(65, 65 + 16)],
        "Birth_Month": [
            "December",
            "January",
            "February",
            "February",
            "March",
            "April",
            "May",
            "June",
            "December",
            "January",
            "February",
            "March",
            "April",
            "May",
            "June",
            "October",
        ],
        "Gender": np.random.choice(["M", "F"], 16),
        "PassDivision": np.random.choice(["I", "II", "III"], 16),
    }
)
```

```python
df
```

```python
pd.get_dummies(df, columns=["Gender", "PassDivision"])
```

```python
df["Birth_Month"] = pd.Categorical(
    df["Birth_Month"],
    ordered=True,
    categories=[
        "January",
        "February",
        "March",
        "April",
        "May",
        "June",
        "July",
        "August",
        "September",
        "October",
        "November",
        "December",
    ],
)
```

```python
df.sort_values(by = "Birth_Month")
```
